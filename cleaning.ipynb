```python
import pandas as pd
```

```python
best_picture_file = "Output/best_picture_omdb_dataframe.csv"
best_director_file = "Output/best_director_omdb_dataframe.csv"
best_adapted_file = "Output/best_adapted_omdb_dataframe.csv"
best_original_file = "Output/best_original_omdb_dataframe.csv"
best_effects_file = "Output/best_effects_omdb_dataframe.csv"
```

```python
best_picture_df = pd.read_csv(best_picture_file)
best_director_df = pd.read_csv(best_director_file)
best_effects_df = pd.read_csv(best_effects_file)
best_original_df = pd.read_csv(best_original_file)
best_adapted_df = pd.read_csv(best_adapted_file)
```

```python
best_picture_df = best_picture_df.drop(["Released", "Ratings", "DVD", "Website", "totalSeasons", "Type", "Poster", "Country", "Plot", "Language", "BoxOffice", "Error", "Response", "Unnamed: 0"], axis = 1)
best_director_df = best_director_df.drop(["Released", "Ratings", "DVD", "Website", "totalSeasons", "Type", "Poster", "Country", "Plot", "Language", "BoxOffice", "Error", "Response", "Unnamed: 0"], axis = 1)
best_original_df = best_original_df.drop(["Released", "Ratings", "DVD", "Website", "Type", "Poster", "Country", "Plot", "Language", "BoxOffice", "Error", "Response", "Unnamed: 0"], axis = 1)
best_adapted_df = best_adapted_df.drop(["Released", "Ratings", "DVD", "Website", "Type", "Poster", "Country", "Plot", "Language", "BoxOffice", "Response", "Unnamed: 0"], axis = 1)
best_effects_df = best_effects_df.drop(["Released", "Ratings", "DVD", "Website", "Type", "Poster", "Country", "Plot", "Language", "BoxOffice", "Error", "Response", "Unnamed: 0"], axis = 1)
```

```python
best_picture_df = best_picture_df.dropna(how = "any", subset=["Awards"])
best_director_df = best_director_df.dropna(how = "any", subset=["Awards"])
best_original_df = best_original_df.dropna(how = "any", subset=["Awards"])
best_adapted_df = best_adapted_df.dropna(how = "any", subset=["Awards"])
best_effects_df = best_effects_df.dropna(how = "any", subset=["Awards"])
```

```python
best_picture_df = best_picture_df.reset_index(drop = True)
best_director_df = best_director_df.reset_index(drop = True)
best_original_df = best_original_df.reset_index(drop = True)
best_adapted_df = best_adapted_df.reset_index(drop = True)
best_effects_df = best_effects_df.reset_index(drop = True)
```

```python
best_director_df[["Director 1", "Director 2", "Director 3", "Director 4"]] = best_director_df.Director.str.split(",", expand = True)
best_director_df[["Production 1", "Production 2", "Production 3","Production 4", "Production 5", "Production 6", "Production 7"]] = best_director_df.Production.str.split(",", expand = True)
best_director_df.to_csv("Output/cleaned_director.csv")
```
