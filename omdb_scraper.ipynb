```python
import json
from config import api_key
import requests
import pandas as pd
import numpy as np
from pandas import json_normalize
```

```python
file = "Resources/data_csv.csv"
df = pd.read_csv(file)
df.head()
```

```python
url = "http://www.omdbapi.com/?apikey=" + api_key + "&t="
bpresponses = []
bactorresponses = []
bactressresponses = []
bdirectorresponses = []
boriginalresponses = []
badaptedresponses = []
beffectsresponses = []
```

```python
df["category"].unique()
```

```python
#searches for best actor in leading role
best_actor = df.loc[(df["category"] == "ACTOR") |
                   (df["category"] == "ACTOR IN A LEADING ROLE")]
#best_actress search
best_actress = df.loc[(["category"] == "ACTRESS") |
                     (df["category"] == "ACTRESS IN A LEADING ROLE")]
# searches for all Best Picture nominees using various names for the award
best_picture = df.loc[(df["category"] == "OUTSTANDING PICTURE") | 
                     (df["category"] == "OUTSTANDING PRODUCTION") |
                     (df["category"] == "OUTSTANDING MOTION PICTURE") |
                     (df["category"] == "BEST MOTION PICTURE") |
                     (df["category"] == "BEST PICTURE")]
# search for best original screenplays
best_original = df.loc[(["category"] == "WRITING (Original Story)") |
                     (df["category"] == "WRITING (Original Screenplay)") |
                     (df["category"] == "WRITING (Screenplay--Original)") |
                     (df["category"] == "WRITING (Story and Screenplay--written directly for the screen)") |
                     (df["category"] == "WRITING (Story and Screenplay--based on material not previously published or produced)")|
                     (df["category"] == "WRITING (Screenplay Written Directly for the Screen)")]
#search for best adapted screenplay
best_adapted = df.loc[(["category"] == "WRITING (Adaptation)") |
                     (df["category"] == "WRITING (Screenplay--Adapted)") |
                     (df["category"] == "WRITING (Screenplay--based on material from another medium)") |
                     (df["category"] == "WRITING (Screenplay Adapted from Other Material)") |
                     (df["category"] == "WRITING (Screenplay Based on Material Previously Produced or Published)")]
#search for best director 
best_director = df.loc[(["category"] == "DIRECTING (Comedy Picture)") |
                     (df["category"] == "DIRECTING (Dramatic Picture)") |
                     (df["category"] == "DIRECTING")]
#search for best special effects
best_effects = best_original = df.loc[(["category"] == "ENGINEERING EFFECTS") |
                     (df["category"] == "SPECIAL EFFECTS") |
                     (df["category"] == "SPECIAL VISUAL EFFECTS") |
                     (df["category"] == "SPECIAL ACHIEVEMENT AWARD (Visual Effects)") |
                     (df["category"] == "VISUAL EFFECTS")]
```

```python
# creates series of movie titles
bptitles = best_picture["entity"]
best_actor_titles = best_actor["entity"]
best_actress_titles = best_actress["entity"]
best_original_titles = best_original["entity"]
best_adapted_titles = best_adapted["entity"]
best_director_titles = best_director["entity"]
best_effects_titles = best_effects["entity"]
```

```python
#best picture scrape
for title in bptitles:
    best_picture_data = requests.get(url + title).json()
    bpresponses.append(best_picture_data)
```

```python
#best actor scrape
for title in best_actor_titles:
    best_actor_data = requests.get(url + title).json()
    bactorresponses.append(best_actor_data)
```

```python
best_actor_omdb_df = json_normalize(bactorresponses)
best_actor_omdb_df.to_csv('Output/best_actor_omdb_dataframe.csv')
```

```python
#best actress scrape
for title in best_actress_titles:
    best_actress_data = requests.get(url + title).json()
    bactressresponses.append(best_actress_data)
best_actress_omdb_df = json_normalize(bactressresponses)
best_actress_omdb_df.to_csv('Output/best_actress_omdb_dataframe.csv')
```

```python
#best director scrape
for title in best_director_titles:
    best_director_data = requests.get(url + title).json()
    bdirectorresponses.append(best_director_data)
best_director_omdb_df = json_normalize(bdirectorresponses)
best_director_omdb_df.to_csv('Output/best_director_omdb_dataframe.csv')
```

```python
#best original scrape
for title in best_original_titles:
    best_original_data = requests.get(url + title).json()
    boriginalresponses.append(best_original_data)
best_original_omdb_df = json_normalize(boriginalresponses)
best_original_omdb_df.to_csv('Output/best_original_omdb_dataframe.csv')
```

```python
#best adapted scrape
for title in best_original_titles:
    best_adapted_data = requests.get(url + title).json()
    badaptedresponses.append(best_adapted_data)
best_adapted_omdb_df = json_normalize(badaptedresponses)
best_adapted_omdb_df.to_csv('Output/best_adapted_omdb_dataframe.csv')
```

```python
#best effects scrape
for title in best_effects_titles:
    best_effects_data = requests.get(url + title).json()
    beffectsresponses.append(best_effects_data)
best_effects_omdb_df = json_normalize(beffectsresponses)
best_effects_omdb_df.to_csv('Output/best_effects_omdb_dataframe.csv')
```
